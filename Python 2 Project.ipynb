```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r"C:\Users\Deena Johnson\Documents\`Data Files\housing.csv")
```

```python
#  1 Create a scatter plot that shows land size on the x-axis and price on the y-axis. Change the color of the points to reflect the type of the property

# Create a scatter plot using Seaborn with 'hue' to differentiate points by 'Type'
plt.figure(figsize=(10, 6))

# Use 'hue' for color differentiation based on 'Type', remove the 'color' parameter
sns.scatterplot(data=df, x='Landsize', y='Price', hue='Type', s=100, palette='viridis')

# Add titles and labels
plt.title('Scatter Plot of Landsize vs. Price by Type')
plt.xlabel('Landsize')
plt.ylabel('Price')

# Display the plot
plt.show()
```

```python
#  3 Create a line chart that shows the year built along the x-axis and the average price per year on the y-axis. You will need to use aggregation and group by to accomplish this
df.groupby('YearBuilt')['Price'].mean()
```

```python
# Calculate the average price per year built
avg_price = df.groupby('Regionname')['Price'].mean().reset_index()

# Set the plot style
sns.set(style="whitegrid")

# Create a line plot
plt.figure(figsize=(14, 7))  # Adjust figure size as needed
sns.barplot(data=avg_price, x='Regionname', y='Price')

# Set plot labels and title
plt.xlabel('Region')
plt.ylabel('Average Price')
plt.title('Average Price per Region')

# Use sns.regplot to add a scatter plot with a regression line (trendline)
# sns.regplot(data=avg_price, x='Regionname', y='Price', marker='o', scatter_kws={'s': 100}, line_kws={'color': 'green'})

# Display the plot with adjusted layout
plt.tight_layout()
plt.show()
```

```python
#  5 Create a bar chart that shows the region on the x-axis and the average price of the property on the y-axis. You will need to use aggregation and group by to accomplish this.

# Calculate the average price per year built
avg_price = df.groupby('YearBuilt')['Price'].mean().reset_index()

# Set the plot style
sns.set(style="whitegrid")

# Create a line plot
plt.figure(figsize=(14, 7))  # Adjust figure size as needed
sns.lineplot(data=avg_price, x='YearBuilt', y='Price', marker='o')

# Set plot labels and title
plt.xlabel('Year Built')
plt.ylabel('Average Price')
plt.title('Average Price per Year Built')

# Use sns.regplot to add a scatter plot with a regression line (trendline)
sns.regplot(data=avg_price, x='YearBuilt', y='Price', marker='o', scatter_kws={'s': 100}, line_kws={'color': 'red'})

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot with adjusted layout
plt.tight_layout()
plt.show()
```

```python
#  7 Create a histogram of property prices
count_home = df['Price'].count()

print(count_home)

plt.figure(figsize=(10,6))
sns.histplot(data=df, x='YearBuilt', bins=len(df['YearBuilt'].unique()), discrete=True, color='green')

plt.title('Number of Homes per Year')
plt.xlabel('Year Built')
plt.ylabel('Price')

plt.show()
```

```python
#  9 Create a box plot that shows the years that each property was built on the x-axis and the region on the y-axis. These box plots should be oriented horizontally

# year_built = df.groupby('YearBuilt')['Propertycount'].count().reset_index()
# region_name = df.groupby('Regionname')['Propertycount'].count().reset_index()

# Set the plot style
sns.set(style="whitegrid")

# Create a bar plot
plt.figure(figsize=(14, 7))  # Adjust figure size as needed
sns.boxplot(data=df, x='YearBuilt', y='Regionname', orient='h', palette='viridis')

# Set plot labels and title
plt.xlabel('Year')
plt.ylabel('Region')
plt.title('Year Built by Region')

# Rotate x-axis labels for better readability if necessary
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```
